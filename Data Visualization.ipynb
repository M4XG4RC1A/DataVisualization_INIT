# Data Visualization app

```python
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns 
```

## Import Data

```python
data = pd.read_csv("water_potability.csv")
nVariables = data.columns.values
```

## Visualization

### Box Plot

```python
bp = []
for n in nVariables:
    data[n].plot(kind='box', title=n)
    plt.show()
```

### Histogram

```python
for n in nVariables:
    data[n].plot(kind='hist', title=n)
    plt.show()
```

### Correlation Map

```python
ax = sns.heatmap(
    data.corr(),
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
    annot=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
plt.show()
```
